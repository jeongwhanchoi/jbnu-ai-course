### Linear Plot

The code `plt.plot([1, 2, 3])` is same as below:
``` python
plt.plot([0,1,2],[1, 2, 3])
```

````python
plt.plot([1, 2, 3])
plt.plot([1,2,3],[2,4,6])
plt.axis([0,4,0,7])
plt.show()
````

### Polynomial Plot

````python
x = np.arange(0.0, 5.0, 0.01)
plt.plot(x, 2**x, 'r--')
plt.plot(x, x**2, 'b')

plt.axis([0,6,0,40])
plt.show()
````

### Scatter Plot

````python
x1=[1,2,11,20,6,9]
y1=[3,7,10,20,16,30]
x2=[5,7,1,6,9,2]
y2=[3,8,10,23,4,32]

plt.title('two scatter')
plt.scatter(x1, y1, marker='s', c='r')
plt.scatter(x2, y2, marker='*', c='b')
plt.xlabel("X data")
plt.ylabel("Y data")
plt.legend(['1','2'])
plt.grid(True)
plt.show()
````

### Bar Chart

````python
x1=[1,3,5,7,9] 
y1=[5,3,9,2,8]
x2=[2,4,6,8,10] 
y2=[8,7,2,11,4]
plt.bar(x1, y1, label = "Set 1", color = "b") 
plt.bar(x2, y2, label = "Set 2", color = "r")
plt.legend()
plt.title("Bar Chart")
plt.show()
````

### Histogram

````python
df = pd.read_csv('../input/wage_conv_store.csv') 
plt.hist(df.hourly_wage, bins=10)
plt.show()
````

### [Box Plot](https://en.wikipedia.org/wiki/Box_plot)

If you have difficulty seeing much of your data, you can use it to quickly identify the range and median of your dataset.

![Box Plot](https://pro.arcgis.com/en/pro-app/help/analysis/geoprocessing/charts/GUID-0E2C3730-C535-40CD-8152-80D794A996A7-web.png)


- **Outlier**: more(less) than 3/2 times of upper(lower) quartile
- **Maximum**: greatest value, excluding outliers
- **Upper Quartile**: 25% of data greater than this value
- **Median**: 50% of data is greater than this value; middle of dataset
- **Lower Quartile**: 25% of data less than this value
- **Minimum**: least value, excluding outliers

````python
df.boxplot(column='hourly_wage', return_type='dict')
plt.show()
````
